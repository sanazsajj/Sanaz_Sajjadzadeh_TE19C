```python
#1
import pandas as pd

df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1)

print(df)
```

```python
#a
true_df = df[df["sysselsï¿½ttning"] == "fï¿½rvï¿½rvsarbetande"]
summa_df = true_df["2019"].tolist()

summa = sum(summa_df)

print(true_df)
print(f"{summa} förvärvsarbetade i Sverige 2019.")
```

```python
#b
false_df = df[df["sysselsï¿½ttning"] == "ej fï¿½rvï¿½rvsarbetande"]
summa2_df = false_df["2019"].tolist()

summa2 = sum(summa2_df)

print(false_df)
print(f"{summa2} förvärvsarbetade inte i Sverige 2019.")
```

```python
#c
import matplotlib.pyplot as plt
color = ["lightpink", "lightblue"]
andel = [summa, summa2]

labels1 = "Förvärvsarbetande","Ej Förvärvsarbetande"

plt.pie(andel, labels = labels1, colors=color, autopct="%1.1f%%")
plt.title("Förvärvsarbetande/Ej Förvärvsarbetande 2019")
plt.axis("Equal")
plt.show()
```

```python
#2a
män_df = df[df["kï¿½n"] == "mï¿½n"]
summa_män_df = män_df.head(6)["2019"].tolist()
summa_män = sum(summa_män_df)

print(män_df.head(6))
print(f"{summa_män} män förvärvsarbetade i Sverige 2019.")
```

```python
#b
kvinnor_df = df[df["kï¿½n"] == "kvinnor"]
summa_kvinnor_df = kvinnor_df.head(6)["2019"].tolist()
summa_kvinnor = sum(summa_kvinnor_df)

print(kvinnor_df.head(6))
print(f"{summa_kvinnor} kvinnor förvärvsarbetade i Sverige 2019.")
```

```python
#c
män2_df = df[df["kï¿½n"] == "mï¿½n"]
summa2_män_df = män_df.iloc[6:12]["2019"].tolist()
summa2_män = sum(summa2_män_df)

print(män_df.iloc[6:12])
print(f"{summa2_män} män förvärvarbetade inte i Sverige 2019.")
```

```python
#d
kvinnor2_df = df[df["kï¿½n"] == "kvinnor"]
summa2_kvinnor_df = kvinnor_df.iloc[6:12]["2019"].tolist()
summa2_kvinnor = sum(summa2_kvinnor_df)

print(kvinnor_df.iloc[6:12])
print(f"{summa2_kvinnor} kvinnor förvärvarbetade inte i Sverige 2019.")
```

```python
import matplotlib.pyplot as plt

color = ["lightpink", "lightblue"]
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
label = ["Män", "män", "Kvinnor", "kvinnor"]

andel = [summa_män, summa2_män, summa_kvinnor, summa2_kvinnor]
ax.bar(label,andel,color=color)
plt.title("Förvärvsarbetande/Ej Förvärvsarbetande 2019")
plt.ylabel("Antal miljoner")
plt.legend(labels=["Förvärvsarbetande","Ej förvärvsarbetande"])
plt.show()
print("Ej förvärvsarbetande = ljusblå")
```
