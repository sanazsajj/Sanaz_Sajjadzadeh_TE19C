```python
#1a
import random

tärningskast = []

for i in range(10):
    i = random.randint(1, 6)
    tärningskast.append(i)

with open("diceRoll.txt", "w") as f:
        f.write(f"Simulera 10 tärningskast: {tärningskast}\n")

#1b
tärningskast.sort()

with open("diceRoll.txt", "a") as f:
    f.write(f"Kastet sorterat: {tärningskast}\n")

#1c
antal_femmor = 0

for i in range(10):
    if tärningskast[i] == 5:
        antal_femmor += 1

with open("diceRoll.txt", "a") as f:
    f.write(f"Antal femmor: {antal_femmor}")
```

```python
#2b
namnlista = []

with open("Provresultat.txt", "r") as f1, open("Provresultat sorterat.txt", "w") as f2:
    for rad in f1:
        namnlista.append(rad)

    namnlista.sort() 

    for i in range(len(namnlista)):
        rad = rad.strip(" ")
        f2.write(namnlista[i])
```

```python
#2c
A = []
B = []
C = []
D = []
E = []
F = []

with open("Provresultat.txt", "r") as f1, open("Provresultat betyg.txt", "w") as f3:
    for rad in f1:
        betyg = int(rad[-3:-1])

        if betyg < 20:
            F.append(rad)

        elif betyg < 30:
            E.append(rad)
            
        elif betyg < 40:
            D.append(rad)

        elif betyg < 50:
            C.append(rad)

        elif betyg < 60:
            B.append(rad)

        elif betyg <= 70:
            A.append(rad)

    f3.write(f"F:\n{F}\n")
    f3.write(f"E:\n{E}\n")
    f3.write(f"D:\n{D}\n")
    f3.write(f"C:\n{C}\n")
    f3.write(f"B:\n{B}\n")
    f3.write(f"A:\n{A}\n")
```
