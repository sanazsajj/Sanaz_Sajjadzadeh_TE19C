```python
#3
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,11)

def f(x):
    y = 3*x+1
    return y

plt.plot(x,f(x))
plt.title("Graf till f(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
