```python
import pandas as pd
import plotly.express as px

#Läser in filen som behövs
ålder_df = pd.read_csv("National_Total_Deaths_by_Age_Group.csv")

fig = px.bar(ålder_df, x="Age_Group", y="Total_Cases") #Skapar ett stapeldiagram där x-axeln är åldersgrupperna och y-axlen är antalet sjuka
fig.update_layout(
    title="Antal sjuka för varje åldersgrupp", #Ger grafen en titel
    xaxis_title="Åldersgrupper", #Ger x-axeln ett namn
    yaxis_title="Antal sjuka" #Ger y-axeln ett namn
)
fig.update_traces(
    marker_color="pink") #Ändrar staplarnas färger
fig.show()
```

```python
import pandas as pd
import plotly.express as px

#Läser in filen som behövs
sverige_df = pd.read_csv("Regional_Daily_Cases.csv")

fig = px.line(sverige_df, x="Date", y="Sweden_Total_Daily_Cases") #Skapar ett linjediagram där x-axeln är datum och y-axeln är antalet sjuka för de datumen
fig.update_layout(
    title="Antal sjuka varje dag i Sverige", #Ger grafen en titel
    xaxis_title="Datum", #Ger x-axeln ett namn
    yaxis_title="Antal sjuka" #Ger y-axeln ett namn
)
fig.update_traces(
    line_color="purple") #Ändrar färg på diagrammet
fig.show()
```

```python
import pandas as pd
import plotly.express as px

#Läser in filen som behövs
ålder_df = pd.read_csv("National_Total_Deaths_by_Age_Group.csv")

antal = [ålder_df["Total_ICU_Admissions"].sum(), ålder_df["Total_Deaths"].sum()]
labels = ["Kritisk vård", "Antal döda"] 
fig = px.bar(ålder_df, x=antal, y=labels) #Skapar ett stapeldiagram där x-axeln är antalet människor och y-axeln är de som dött och de som behövde kritisk vård
fig.update_layout(
    title="Behövde kritisk vård/Antal döda", #Ger grafen en titel
    xaxis_title="Antal människor", #Ger x-axeln ett namn
    yaxis_title = "" #Tar bort y-axelns namn
)
fig.update_traces(
    marker_color=["purple", "pink"] #Ändrar staplarnas färger
    )
fig.show()
```

```python
import pandas as pd
import plotly.express as px

#Läser in filen som behövs
ålder_df = pd.read_csv("National_Total_Deaths_by_Age_Group.csv")

antal = [ålder_df.head(6)["Total_Deaths"].sum(), ålder_df.iloc[6:10]["Total_Deaths"].sum()]
labels = ["Antal döda inte i riskgruppen", "Antal döda i riskgruppen"]

#Skapar ett cirkelndiagram med hur många som dött i riskgruppen och antalet som dött inte i riskgruppen
fig = px.pie(ålder_df, values=antal, names=labels, color=labels, color_discrete_map={"Antal döda inte i riskgruppen":"pink",
"Antal döda i riskgruppen":"purple"}) #Ändrar färg på delarna

fig.update_layout(
    title="Antal döda i riskgruppen/inte i riskgruppen", #Ger diagrammet en titel
)
fig.show()
```
