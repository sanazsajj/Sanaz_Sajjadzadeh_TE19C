```python
#(e)
import math
import random
Innanför = 0
for i in range (20):
    x = random.uniform (-1,1)
    y = random.uniform (-1,1)
    d = math.sqrt(x**2+y**2) #Avståndet från origo till de slumpmässiga punkterna.
    print(d)
    if d <= 1:
        Innanför += 1 
        print("Innanför")
    else:
        print("Utanför")
    print()
    a = Innanför/20
print(f"Antalet punkter inuti cirkeln delat med antalet punkter(20) är {a}.")
```

```python
#(f)
import math
import random
Innanför = 0
for i in range (10000):
    x = random.uniform (-1,1)
    y = random.uniform (-1,1)
    d = math.sqrt(x**2+y**2)
    if d <= 1:
        Innanför += 1 

    a = 4*(Innanför/10000)
print(a) #Ju mer punkter det är desto mer av cirkelns area täcker de och därför blir talet nära pi eftersom cirkelns area är pi. Man delar med 4 eftersom innan delar man punkterna i cirkeln(cirkelns area) med alla punkter (kvadratens area) och då får man cirkelns andel av kvadraten. Kvadratens area är 4 och cirkelns area är pi, å det blir pi/4, och då multiplicerar man med 4 för att få cirkelns area igen.
```

```python
#(g)
import math
import random
import matplotlib.pyplot as plt
for i in range (1000):
    x = random.uniform (-1,1)
    y = random.uniform (-1,1)
    d = math.sqrt(x**2+y**2) #Avståndet från origo till de slumpmässiga punkterna.
    if d <= 1:
        plt.plot(x, y, "b*") #Blå punkter är inuti cirkeln.
    else:
        plt.plot(x, y, "r*") #Röda punkter är utanför.
```

```python
#(e)
x = int(input("Ange ett starttal:"))
y = int(input("Ange ett sluttal"))
a = int(input("Ange ett tal:"))
b = int(input("Ange ett till tal:"))
Burr = 0
Birr = 0
Burr_Birr = 0
for i in range (x,y):
    if i%a == 0 and i%b == 0:
        Burr_Birr += 1
    elif i%a == 0:
        Burr += 1
    elif i%b == 0:
        Birr += 1
print(f"Antal Burr,Birr = {Burr_Birr}")
print(f"Antal Burr = {Burr}")
print(f"Antal Birr = {Birr}")       
```

```python
#(f) Låta användaren mata in start, slut, multipel för "Burr", multipel för "Birr" och ett multipel för "Barr".
x = int(input("Ange ett starttal:"))
y = int(input("Ange ett sluttal"))
a = int(input("Ange ett tal:"))
b = int(input("Ange ett till tal:"))
c = int(input("Ange ett till tal:"))
for i in range (x,y):
    if i%a == 0 and i%b == 0:
        print("Burr,Birr", end=" ")
    elif i%a == 0 and i%c == 0:
        print("Burr,Barr", end=" ")
    elif i%b == 0 and i%c == 0:
        print("Birr,Barr", end=" ")
    elif i%a == 0 and i%b == 0 and i%c == 0:
        print("Burr,Birr,Barr", end=" ")
    elif i%a == 0:
        print("Burr", end=" ")
    elif i%b == 0:
        print("Birr", end=" ")
    elif i%c == 0:
        print("Barr", end=" ")
    else:
        print(i, end=" ")
```
